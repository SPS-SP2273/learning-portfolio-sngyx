```python
# I assumed this task was just changing the file name str, changing the actual
# file name shall happen later
# I'll change it if that's not the case!

def rename_my_file(path: str):
    '''Rename a file from month-date-city.txt to mm-dd-city.txt'''
    month_key = {
                'jan': '01', 'feb': '02', 'mar': '03',
                'apr': '04', 'may': '05', 'jun': '06',
                'jul': '07', 'aug': '08', 'sep': '09',
                'oct': '10', 'nov': '11', 'dec': '12'
                }
    try:
        filename = path.split(os.path.sep)[-1] # take only file name from path
        # assuming letter case won't be an issue
        for month_name, month_num in month_key.items(): 
            filename = filename.replace(month_name, month_num)
        return filename
    except Exception as e:
        print(f'An error occured. \n Error: \n {e}')
    
rename_my_file('hihi/testtesthellooo/oct-08_singapore.txt')
```

```python
renameThese_list = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for old_filename in renameThese_list:
    try:
        #find full old name
        full_old_filenameSearch = glob.glob(
                                    os.path.join(
                                            '*os-collaboration-exercise-data', 
                                            old_filename))
        if len(full_old_filenameSearch) > 0:
            # rename all duplicates too
            for full_old_filename in full_old_filenameSearch:
                # make new name
                new_filename = rename_my_file(old_filename)
                full_new_filename = os.path.join(
                                            'os-collaboration-exercise-data', 
                                            new_filename)
                # rename the actual file
                os.rename(full_old_filename, full_new_filename) 
                # print progress
                print(f'Changed: {old_filename} ----> {new_filename}') 
        elif len(full_old_filenameSearch) == 0:
              print(f'No such file ({old_filename}) found.')   
    except Exception as e:
        print(f'An error occured. \n Error: \n {e}')
```

## Task 7


```python
# first remove extra Cairos
extra_cairo_list = glob.glob(os.path.join('*','*cairo*'))
for extra_cairo in extra_cairo_list:
    os.remove(extra_cairo)

# now move everything
everything_list = glob.glob(os.path.join('os-collaboration-exercise-data', 
                                         '*.*'))
for data in everything_list:
    for city in city_list:
        if city.lower() in data.lower():
            filename = data.split(os.path.sep)[-1] # take file name from path
            new_filepath = os.path.join('os-collaboration-exercise-data', 
                                        city,
                                        filename)
            shutil.move(data, new_filepath) 
            print(f'Moved {data} to {new_filepath}.')
        
```
