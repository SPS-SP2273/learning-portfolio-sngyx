## Code Block 1:

```python
import timeit
time = timeit.timeit('squares = [i**2 for i in range(10)]')

squares = [i**2 for i in range(10)]

print(time)
print(squares)
```

### Result:
Time taken: 0.39415283300331794 seconds, \
squares = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

## Code Block 2:

```python
import timeit
time = timeit.timeit(stmt='for i in range(10):; squares.append(i**2)', setup='squares = list()')

squares = list()
for i in range(10):
    squares.append(i**2)
    
print(time)
print (squares)
```
